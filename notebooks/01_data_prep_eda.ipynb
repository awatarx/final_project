```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (going up one level from the notebooks directory)
df = pd.read_csv('../data/heart.csv')
df.head()
```

```python
# Check for class imbalance
df.info()

print(df['target'].value_counts())

sns.countplot(x='target', data=df)
plt.title('Target Variable Distribution (1 = Disease, 0 = No Disease)')
plt.show()
```

```python
# Correlation matrix to see which features impact the target most
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 1. Histograms for all features
print("--- Feature Distributions ---")
df.hist(figsize=(15, 12), bins=15, color='steelblue', edgecolor='black')
plt.tight_layout()
plt.show()

# 2. Boxplots comparing key features against Heart Disease (Target)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Updated to include hue="target" and legend=False to fix the FutureWarning
sns.boxplot(ax=axes[0], x="target", y="age", hue="target", data=df, palette='Set2', legend=False)
axes[0].set_title("Age vs Heart Disease")

sns.boxplot(ax=axes[1], x="target", y="cholestoral", hue="target", data=df, palette='Set2', legend=False)
axes[1].set_title("Cholesterol vs Heart Disease")

sns.boxplot(ax=axes[2], x="target", y="max_hr", hue="target", data=df, palette='Set2', legend=False)
axes[2].set_title("Max Heart Rate vs Heart Disease")

plt.tight_layout()
plt.show()
```
